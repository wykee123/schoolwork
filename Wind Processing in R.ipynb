```r
data=read.csv("alldir_allspeed.csv")
```

```r
data=data.frame(data)
```

```r
data=subset(data, hourly_mps!=0 & hourly_mps<=30)
```

```r
data$direction[data$direction == 360] = 0
```

```r
summary(data)
```

```r
hist(data$hourly_mps,breaks=seq(0,max(data$hourly_mps)+0.5,0.5),main='Histogram of wind speed')
```

```r
quantile(data$hourly_mps,probs=c(0,0.5,0.75,0.80,0.90,0.95,0.98,0.99,0.999,1))
```

```r
hist(data$direction,breaks=seq(0,max(data$direction)+10,10),main='Histogram of wind direction')
```

```r
quantile(data$direction,probs=c(0,0.5,0.75,0.80,0.90,0.95,0.98,0.99,0.999,1))
```

```r
plot(data$direction,data$hourly_mps,xlab='wind direction',ylab='wind speed',main='Wind speed - direction scatter plot')
```
