```python
import pandas as pd
import numpy as np
```

```python
windraw=pd.read_csv("OTBD (4).csv")
```

```python
windraw.head(10)
```

```python
windraw = pd.DataFrame({"station":windraw.station,"date_time":windraw.valid,"direction":windraw.drct,"hourly_knt":windraw.sknt,"gust_knt":windraw.gust})
```

```python
windraw.head(10)
```

```python
windraw["hourly_mps"]=windraw.hourly_knt*0.514444
```

```python
windraw.describe()
```

```python
windraw = windraw[windraw['direction'].notna()]
windraw.describe()
```

```python
windraw = windraw[windraw['hourly_knt'].notna()]
windraw.describe()
```

```python
windraw=windraw[windraw.hourly_mps<30]
windraw.describe()
```

```python
windraw.head(100)
```

```python
windraw.tail(100)
```

```python
windraw.describe()
```

```python
windraw["gust_mps"]=windraw.gust_knt*0.514444
```

```python
windraw.head()
```

```python
windrawgg=windraw[windraw.gust_mps>windraw.hourly_mps]
```

```python
windrawgg.describe()
```

```python
windrawgg["gust_factor"]=windrawgg["gust_mps"]/windrawgg["hourly_mps"]
```

```python
windrawgg.describe()
```

```python
windrawgg[['year', 'month','day']] = windrawgg['date_time'].str.split('-', expand=True)
```

```python
windrawgg.head()
```

```python
windrawgg[['date','time']] = windrawgg['day'].str.split(' ', expand=True)
```

```python
windrawgg.head()
```

```python
windrawgg=windrawgg.drop(columns=['day'])
```

```python
windrawgg.head()
```

```python
windrawgg=windrawgg.drop(columns=['date_time'])
```

```python
windrawgg.head()
```

```python
windrawgg.describe()
```

```python
windrawgg['dir']=windrawgg['direction'].replace(360,0)
```

```python
windrawgg.describe()
```
